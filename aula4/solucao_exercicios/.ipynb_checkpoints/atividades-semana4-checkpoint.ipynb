# Atividade 1
Construa um gerador de trens Poissonianos de potenciais de ação. Na literatura, isso usualmente significa a realização de um processo de Poisson homogêneo (ou estacionário), ou seja, com taxa $\lambda$ constante: a probabilide de ocorrer um disparo em um intervalo de tempo $\Delta t$ é

\begin{equation}
P_D(t,t+\Delta t) = \lambda \Delta t
\end{equation}

```python
import random
import matplotlib.pyplot as plt

def homogenous_poisson(rate,delta_T,total_time):
    spike_times = []
    spike_criterium = rate*delta_T
    current_time = 0
    
    while current_time <= total_time:
        chance = random.random()
        if spike_criterium > chance:
            spike_times.append(current_time)
        current_time += delta_T
    return spike_times

def isi_estimate(spike_times):
    isi = []
    for position in range(len(spike_times)-1):
        current_isi = spike_times[position+1] - spike_times[position]
        isi.append(current_isi)
    return isi
```

#### Outro método: gerando os intervalos interspike diretamente através de uma distribuição exponencial

Podemos produzir uma distribuição exponencial, necessária para esse método, através de uma variável aleatória $x$ uniformemente distribuída entre 0 e 1 através do método da transformada inversa para amostragem.

Sendo $F(X = x_j) = p_j$, $\int p_j dj = 1$ a função densidade de probabilidade da variável aleatória que queremos produzir, podemos gerar um número distribuído de acordo com ela, $x_j$, fazendo

\begin{equation}
X = x_j \iff p(x_{j-1}) \leq U \leq p(x_j)
\end{equation}

Para $p_{t_0} = e^{-\lambda t_0}$, $t_1 = t_0 + \Delta t$:

\begin{equation}
e^{-\lambda t_0} \leq U \leq e^{-\lambda (t_0 + \Delta t)}
\end{equation}

Aplicando o logaritmo natural em ambos lados da desigualdade

\begin{equation}
-\lambda t_0 \leq ln(U) \leq -\lambda (t_0 + \Delta t)
\end{equation}

Dividindo por $-\lambda$ obtemos

\begin{equation}
t_0 \leq -\frac{1}{\lambda} ln(U) \leq t_0 + \Delta t
\end{equation}

Como o posicionamento (no tempo) do primeiro spike, $t_0$, é arbitrário, podemos escolher a situação em que $t_0 = 0$. Assim,

\begin{equation}
0 \leq -\frac{1}{\lambda} ln(U) \leq \Delta t
\end{equation}

Escolhendo oportunamente a situação de igualdade entre os termos não-nulos (isso é um passo ilegal? me parece legítimo), podemos definir $\Delta t$ e calcular valores para o próximo spike $t_{i+1}$ a partir do instante anterior $t_i$ fazendo

\begin{equation}
\Delta t = \frac{-ln(U)}{\lambda} \implies t_{i+1} = t_i + \Bigg[\frac{-ln(U)}{\lambda} \Bigg]
\end{equation}

```python
#Another way of generating spikes according to a poisson distribution (faster, limited to constant firing rate)
#Reasoning described above (pt)

import numpy as np

def homogenous_poisson_fixed_rate(rate,steps):
    isi = []
    last_spike = 0
    
    #generating the ISI distribution
    for i in range(steps):
        delta_T = -np.log(random.random())/rate
        next_spike = last_spike + delta_T
        isi.append(next_spike)
        
    #and the spike times
    spikes_times = []
    t= 0 #t_0
    for interval in isi:
        t += interval
        spikes_times.append(t)
    return isi,spike_times
```

# Atividade 2
Gere um trem de disparos com uma taxa compatível com a de neurônios corticais. Estime então a taxa de disparos instantânea, via 
<ul>
    <li> histograma temporal
    <li> média movel (kernel retangular):
        \begin{equation}
           0 < t < \theta \implies K(t) = \frac{1}{\theta}\\
           \text{else} \quad K(t) = 0
        \end{equation}
    <li> kernel gaussiano centrado no PA (acausal)
        \begin{equation}
            K(t) = \frac{1}{\sigma \sqrt{2}} exp \bigg(- \frac{t^2}{2\sigma^2} \bigg)
        \end{equation}
    <li> kernel tipo alpha-function
        \begin{equation}
           t > 0 \implies K(t) = \frac{t}{\tau^2} exp\bigg(-\frac{t}{\tau} \bigg)\\
           \text{else} \quad K(t) = 0
        \end{equation}
</ul>

Tente escolher parâmetros (largura de janelas / kernel) comparáveis para cada caso. Para tanto, é importante pensar na relação entre os parâmetros θ, σ, τ em cada kernel. Para mais detalhes sobre este tipo de escolha, veja (Shimazaki e Shinomoto 2010).

Uma taxa de disparos compatível com a atividade de neurônios corticais da V camada do córtex visual é de 10-15Hz (Sun & Dan, 2009). Utilizarei esta como minha referência.

Sun & Dan 2009, source:
www.pnas.orgcgidoi10.1073pnas.0903962106

```python
def kernel_conv(kernel,current_time,spike_time,resolution):
    if kernel == 'rectangular':
        temporal_displacement = abs(current_time - spike_time)
        if temporal_displacement < resolution:
            return 1/resolution
        else:
            return 0
    
    elif kernel == 'gaussian':
        temporal_displacement = abs(current_time - spike_time)
        return np.exp(-temporal_displacement**2/(2*resolution**2))/resolution*np.sqrt(2)
    
    elif kernel == 'alpha':
        temporal_displacement = current_time - spike_time
        if temporal_displacement > 0:
            return (temporal_displacement/resolution**2)*np.exp(-temporal_displacement/resolution)
        else:
            return 0
    else:
        raise Exception("Invalid kernel.")

def windowed_signal(spike_times,starting_point=0,sampling_rate=0.01,kernel='rectangular',resolution=0.1):
    '''
    INPUT:
    starting_point: first timepoint [float]
    sampling_rate: 1/# of points in which we shall estimate our firing rate [float]
    kernel: the kernel each spike shall be convolved with (options: rectangular, gaussian and alpha) [string]
    resolution: the parameter controlling the width of the window (i.e. sigma in the gaussian kernel,
        window width in the rectangular kernel) [float]
    spike_times: an array containing the timestamp for each spike [list(float)]
    
    OUTPUT:
    rates: an array containing the estimated spiking rate at each timepoint [list(float)]
    '''
    current_time = starting_point
    sampling_rate = 0.01
    rates = []
    
    while current_time <= max(spike_times):
        current_time += sampling_rate
        contribution_to_rate = []
        '''We calculate the contribution each spike in the train (instant) makes to the firing rate estimated at
        time (current_time) by convolving the spike with a given kernel, considering the temporal displacement
        between them.

        Larger temporal displacements should yield less contribution to the rate estimate, whereas spikes closer
        to the timepoint we are interested in should yield a larger contribution.'''
        for spike_timestamp in spike_times:
            contribution_to_rate.append(kernel_conv(kernel,current_time,spike_timestamp,resolution=resolution))

        #Sum the contributions of all spikes to achieve the rate estimate at time current_time
        current_rate = np.sum(contribution_to_rate)
        #Which we append to our rates list, which will yield our firing rate (hz) x time (s) plot afterwards
        rates.append(current_rate)
    return rates
```

```python
#Cortical (V layer) neurons in the visual cortex: ~10-15Hz. I choose it to be
rate = 15
delta_T = 0.01
total_time = 10

spike_times = homogenous_poisson(rate,delta_T,total_time)
```

```python
plt.figure(figsize=(24,8))
plt.eventplot(spike_times)
plt.xlabel('Time (s)')
plt.show()
```

```python
#Histogram
plt.figure(figsize=(18,6))
plt.hist(spike_times,bins=15) #~0.66s per bin
#too fine: above 70 bins
#too coarse: below 15 bins
```

```python
rates_rec = windowed_signal(spike_times,starting_point=0,sampling_rate=0.01,kernel='rectangular',resolution=0.4)

plt.figure(figsize=(18,6))
plt.plot(rates_rec)
```

```python
rates_gauss = windowed_signal(spike_times,starting_point=0,sampling_rate=0.1,kernel='gaussian',resolution=0.3)

plt.figure(figsize=(18,6))
plt.plot(rates_gauss)
```

```python
rates_alpha = windowed_signal(spike_times,starting_point=0,sampling_rate=0.1,kernel='alpha',resolution=0.3)

plt.figure(figsize=(18,6))
plt.plot(rates_alpha)
```
